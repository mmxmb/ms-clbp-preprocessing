# Stage 3
## Calculate CLBP sign and magnitude components

```python
n_points = 8
radius = 4

clbp_s = local_binary_pattern(img, n_points, radius, 'uniform')
plt.imshow(clbp_s)
plt.title("Sign CLBP component")
```

```python
clbp_m = local_binary_pattern_magnitude(img, n_points, radius)
plt.imshow(clbp_m)
plt.title("Magnitude CLBP component")
```

# Stage 4
### Partition CLBP component matrices into overlapping patches of size 32x32

```python
def show_image_matrix(imgs, cols, rows):
    fig = plt.figure(figsize=(8, 8))
    for i in range(1, cols * rows + 1):
        img = imgs[i-1]
        fig.add_subplot(rows, cols, i)
        plt.imshow(img)
    plt.show()
```

Partition the following 85x85 image into 16 overlapping patches. Each consecutive patch overlaps 1/2 of the previous patch.

```python
scales = [1/3]
(third_scale_img, ) = rescale_img(img, scales)
third_scale_clbp_m = local_binary_pattern_magnitude(third_scale_img, n_points, radius) # 85 x 85 pixels
plt.imshow(third_scale_clbp_m)
```

```python
third_scale_clbp_m_patches = overlapping_patch_partition(third_scale_clbp_m, 32)
show_image_matrix(third_scale_clbp_m_patches, 4, 4)
```

Show example patch.

```python
patch_size = 32
clbp_s_patches = overlapping_patch_partition(clbp_s, patch_size)
plt.imshow(clbp_s_patches[0])
plt.title("Sign CLBP component 32x32 patch")
```

```python
clbp_m_patches = overlapping_patch_partition(clbp_m, patch_size)
plt.imshow(clbp_m_patches[0])
plt.title("Magnitude CLBP component 32x32 patch")
```

# Stage 5
## Calculate occurence histogram for each component patch

```python
n_bins = 16
clbp_s_hist, clbp_s_bins = np.histogram(clbp_s_patches[0], range(n_bins + 1))
plt.hist(clbp_s_patches[0], bins=range(n_bins + 1))
plt.title("Sign CLBP component patch occurence histogram")
plt.show()
```

```python
clbp_m_hist, _ = np.histogram(clbp_m_patches[0], range(n_bins + 1))
plt.hist(clbp_m_patches[0], bins=range(n_bins + 1))
plt.title("Magnitude CLBP component patch occurence histogram")
plt.show()
```

# Stage 6
## Concatenate two histograms to get a single feature vector

```python
hist_feature_vector = np.append(clbp_s_hist, clbp_m_hist)
hist_feature_vector
```
